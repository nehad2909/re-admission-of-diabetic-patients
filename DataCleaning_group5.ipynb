```python
# Removing the columns which are not required

diabetic_data = diabetic_data.drop(axis=1, columns=["citoglipton", "examide", "encounter_id", "payer_code", "weight"])
```

```python
for column in diabetic_data:
    print(column)
    print(diabetic_data[column].value_counts())
```
